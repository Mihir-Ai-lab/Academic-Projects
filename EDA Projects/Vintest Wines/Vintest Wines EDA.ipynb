# **4. Data Acquisition & Description**
---

- For the markets in the U.S Vintest has collected data over the past few years that connects human quality of tasting to laboratory-based physiochemical tests.

- They also intend find a solution for increased control over certification, quality assessment and assurance process though this exercise.

- We are working with a direct source of link to head start our work.

- Features of the data set as per Vintest:

|Id|Feature|Description|
|:--|:--|:--|
|01||fixed acidity| most acids involved with wine or fixed or nonvolatile (do not evaporate readily).| 
|02|volatile acidity|the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste.| 
|03|citric acid|found in small quantities, citric acid can add ‘freshness’ and flavor to wines.| 
|04|residual sugar | the amount of sugar remaining after fermentation stops, it’s rare to find wines with less than 1 gram/liter |
|||and wines with greater than 45 grams/liter are considered sweet.|
|05| chlorides| the amount of salt in the wine.|
|06| free sulfur dioxide  | free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion;|
|||it prevents microbial growth and the oxidation of wine.|
|07|  total sulfur dioxide | amount of free and bound forms of S02; in low concentrations, SO2 is mostly undetectable in wine,|
|||but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine|
|08| density| the density of water is close to that of water depending on the percent alcohol and sugar content|
|09| pH | describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic);|
||| most wines are between 3-4 on the pH scale.|
|10| sulphates| a wine additive which can contribute to sulfur dioxide gas (S02) levels, wich acts as an antimicrobial and antioxidant.|
|11| alcohol| the percent alcohol content of the wine.|
|12| quality| score between 0 to 10.| 


- Vintest is looking for the right ingredients, and we found that fixed acidity is particularly important as it directly influences:

    - The balance and taste of the wine.

    - The growth and vitality of yeast during fermentation and the protection of the wine from bacteria.

- Citric acid is sometimes added to increase the acidity of wine, but this can have a negative effect, as it also increases the growth of unwanted microbes. Typically, one would expect around 500 mg/L of citric acid to be added to the wine after fermentation.

- Total acidity refers to the test that yields the total of all acids present, while the strength of acidity is measured according to pH, with most wines having a pH between 2.9 and 3.9.

- Generally, the lower the pH, the higher the acidity in the wine. There is no direct connection between total acidity and pH, as it is possible to find wines with a high pH but high acidity.

- In wine tasting, the term “acidity” refers to the fresh, tart, and sour attributes of the wine, which are evaluated in relation to how well the acidity balances out the sweetness and bitter components of the wine, such as tannins (which are more predominantly present in red wines).

- Three primary acids are found in wine grapes: tartaric, malic, and citric acids. During the course of winemaking and in the finished wines, acetic, butyric, lactic, and succinic acids can also play significant roles. 

- Most of the acids involved with wine are fixed acids, with the notable exception of acetic acid, which is volatile and mostly found in vinegar and can contribute to the wine fault known as volatile acidity. 

- Sometimes, additional acids, such as ascorbic, sorbic, and sulfurous acids, are used in winemaking.

- Given Vintest's reputation for quality wine, wine fault tolerance will be close to negligible, especially when they are entering a new country.

- On average, volatile acidity in white wines is 0.6 g/L, while in red wine, it is 0.43 g/L.

```python
# Making a copy of the Dataset incase it needs to be reffered back to
df_copy = df_vintest.copy()
```

```python
# Changing quality ratings from numerical to categorical variable

# Wines rate lower than 4 have been categorized as Rejected
# Wines rated between 5-7 have been categorized as Average
# Wines rated between 7-9 have been categorized as Good
# Wines rated 9 & above have been categorized as Excellent

df_copy['quality category'] = df_copy['quality'].map({3: 'rejected', 4: 'rejected', 5: 'average', 6: 'average', 7: 'good', 8: 'good', 9: 'excellent'})
```

```python
# Performing half round ups to alcohol content since the wine labels generally read in rounded values

df_copy['alcohol'] = df_copy['alcohol'].apply(lambda x: round(x * 2) / 2)

# Dropping Duplicated Rows from the copy of the Dataset

print('\n' + '-'*40 + '\n')
print("Number of Rows before de-duplication:", df_copy.shape[0])
df_copy.drop_duplicates(inplace = True)
print('\n' + '-'*40 + '\n')
print("De-Duplication Complete")
print('\n' + '-'*50 + '\n')
print("Number of Rows After Dropping Duplicates:", df_copy.shape[0])
```

```python
print('\n' + '-'*60 + '\n')
print("Creating Box Plots to identify Inconsistant Data in Acidity")
print('\n' + '-'*60 + '\n')

acidity_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'total sulfur dioxide', 'free sulfur dioxide']

# Create a figure with a grid of subplots
fig, axes = plt.subplots(nrows=len(acidity_cols), ncols=2, figsize=(15, len(acidity_cols)*7))

# Loop through each acidity parameter and create a boxplot on the corresponding subplot
for i, col in enumerate(acidity_cols):
    row = i // 2
    col_num = i % 2
    sns.boxplot(x='quality', y=col, data=df_copy, ax=axes[i, 0], palette='rocket')
    axes[i, 0].set_xlabel('Quality', size=14)
    axes[i, 0].set_ylabel(col, size=14)
    axes[i, 0].tick_params(labelsize=12)
    axes[i, 0].set_title('Quality vs ' + col, size=16)
    axes[i, 0].grid(b=True)

    sns.histplot(x=col, data=df_copy, bins=20, ax=axes[i, 1], kde=True, color='red')
    axes[i, 1].set_xlabel(col, size=14)
    axes[i, 1].set_ylabel('Frequency', size=14)
    axes[i, 1].tick_params(labelsize=12)
    axes[i, 1].set_title(col + ' Distribution', size=16)
    axes[i, 1].grid(b=True)

plt.tight_layout()
plt.show()
```

```python
# Identifying the top 5% of values in each acidity parameter

upper_limit = df_copy.quantile(0.95)
acidity_outliers = df_copy[df_copy[acidity_cols].apply(lambda x: x > upper_limit[x.name]).any(axis=1)]

print('\n' + '-'*80 + '\n')
print("Number of Inconsistant Data Points considering top 5% values are incosistant:", acidity_outliers.shape[0])
print('\n' + '-'*80 + '\n')
```

```python
# Dropping Acidity Outliers from the copy of the Dataset

print("Number of Rows before droping:", df_copy.shape[0])
df_copy.drop(acidity_outliers.index, inplace=True)
print('\n' + '-'*60 + '\n')
print("Number of Rows After Dropping Inconsistant Data Points:", df_copy.shape[0])
print('\n' + '-'*60 + '\n')
```

### Observation

- For residual sugar, we intended to categorize them into three different categories, so that we can analyze them separately based on their categories.

- However Since the NUmber of Sweet wines in the Dataset is very low, We decided to remove OUtliers with IQR Method

```python
# Identifying the top 5% of values in sulphates & Chlorides

# Define the other potential outlier columns
other_outlier_cols = ['sulphates', 'chlorides']
upper_limit = df_copy.quantile(0.95)
other_outliers = df_copy[df_copy[other_outlier_cols].apply(lambda x: x > upper_limit[x.name]).any(axis=1)]

print('\n' + '-'*50 + '\n')
print("Top 5% values in Sulphates & Chlorides:", other_outliers.shape[0])
print('\n' + '-'*50 + '\n')
```

```python
# Dropping Other Outliers from the copy of the Dataset

print("Number of Rows before droping:", df_copy.shape[0])
df_copy.drop(other_outliers.index, inplace=True)
print('\n' + '-'*70 + '\n')
print("Number of Rows After Dropping Sulphate & Chloride Outliers:", df_copy.shape[0])
print('\n' + '-'*70 + '\n')
```

```python
# Calculate the IQR for density

Q1 = df_copy['density'].quantile(0.25)
Q3 = df_copy['density'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower limits for density
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filter out the outliers from density
density_outliers = df_copy[(df_copy['density'] < lower_limit) | (df_copy['density'] > upper_limit)]

print("Number of outliers in density using IQR method:", density_outliers.shape[0])
print('\n' + '-'*50 + '\n')
```

```python
# Dropping Density Outliers from the copy of the Dataset

print("Number of Rows before droping:", df_copy.shape[0])
df_copy.drop(density_outliers.index, inplace=True)
print('\n' + '-'*55 + '\n')
print("Number of Rows After Dropping Density Outliers:", df_copy.shape[0])
print('\n' + '-'*55 + '\n')
```

```python
# Calculate the IQR for residual Sugar

Q1 = df_copy['residual sugar'].quantile(0.25)
Q3 = df_copy['residual sugar'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower limits for density
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filter out the outliers from density
residual_sugar_outliers = df_copy[(df_copy['residual sugar'] < lower_limit) | (df_copy['residual sugar'] > upper_limit)]

print("Number of outliers in Residual Sugar using IQR method:", residual_sugar_outliers.shape[0])
print('\n' + '-'*60 + '\n')
```

```python
# Dropping Residual Sugar Outliers from the copy of the Dataset

print("Number of Rows before droping:", df_copy.shape[0])
df_copy.drop(residual_sugar_outliers.index, inplace=True)
print('\n' + '-'*60 + '\n')
print("Number of Rows After Dropping Residual Sugar Outliers:", df_copy.shape[0])
print('\n' + '-'*60 + '\n')
```

```python
# Create a dataframe with all the outliers that were removed
df_outliers = pd.concat([acidity_outliers, other_outliers, density_outliers, residual_sugar_outliers])

# Rename df_copy to df_clean
df_clean = df_copy.copy()

print("Created df_clean with Outlier Removed with rows = ", df_clean.shape[0])
print('\n' + '-'*80 + '\n')
print("Created df_outliers with all Outliers that were removed with rows = ", df_outliers.shape[0])
print('\n' + '-'*80 + '\n')
```

### **8.10 How does the presence of Citric Acid affect other physiochemical properties of wine?**

```python
# Group the dataset by the presence of citric acid and calculate the mean values for each group
citric_present = df_clean.groupby(df_clean['citric acid']>0).mean()

# Create a new dataframe with the desired columns
df_citric = pd.DataFrame(columns=['Feature', 'Citric Present Mean', 'Citric Absent Mean', 'Difference'])

# Loop through the columns in the dataset
for col in citric_present.columns:
    # Calculate the mean values for each group
    mean_citric_present = citric_present.loc[True, col]
    mean_citric_absent = citric_present.loc[False, col]
    
    # Calculate the difference between the mean values
    diff = mean_citric_present - mean_citric_absent
    
    # Append the results to the new dataframe
    df_citric = df_citric.append({'Feature': col,
                                  'Citric Present Mean': mean_citric_present,
                                  'Citric Absent Mean': mean_citric_absent,
                                  'Difference': diff},
                                 ignore_index=True)

# Print the resulting dataframe

print("Difference between the mean Values of Wines that have Citric Acid Present/Absent")
print('\n' + '-'*80 + '\n')
print(df_citric)
```

- Citric acid has a positive correlation with residual sugar, which makes sense as citric acid is often added to wines to balance their sweetness.
- The presence of citric acid in wine seems to have a positive impact on wine quality, as wines with citric acid have a higher average quality score compared to wines without citric acid.
- However, it is important to note that citric acid has a negative correlation with density, which is a measure of the wine's body. This suggests that adding citric acid to wine may result in a lighter body, which may not be desirable for certain types of wine.
